```python
import pandas as pd
df=pd.read_csv("test.csv")
print(df)
from matplotlib import pyplot as plt
plt.style.use('ggplot')
plt.plot(df.Name,df.Number_of_skills ,label="skills",color="red",linewidth=4,linestyle=':',marker='d')
plt.legend()
plt.ylabel("Number_of_skills",fontsize=20,color='green')
plt.xlabel("Name",fontsize=20,color='blue')
plt.title("Skills analysis",fontsize=30,color='orange')
plt.show()
```
